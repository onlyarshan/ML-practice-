```python
#Task 5: Players who have won the most 'Player of the match'
##and team whcih has most wins
print('most player of the match award\n',df_matches['player_of_match'].value_counts().idxmax())#value_counts gives us the occurance of a value in a colimn
print('most team winds\n',df_matches['winner'].value_counts().idxmax())
```

```python
#Task 6: Find teams who have won match by a huge margin
#assume win by runs greater than 100 and win by wickets greater than 8 is the criteria
big_margin=df_matches[(df_matches['win_by_runs']>=100) | (df_matches['win_by_wickets']>=8)]
big_margin['winner'].value_counts()
```
